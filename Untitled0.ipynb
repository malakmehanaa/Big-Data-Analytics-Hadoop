#  Customer Data — Full Preprocessing Pipeline
**Dataset:** 1,000,000 rows × 12 columns  
**Goal:** Production-grade cleaning, validation, and feature engineering — one step per cell.

### Imports & Global Settings

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import re, warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({
    'figure.facecolor': '#FAFAFA',
    'axes.facecolor':   '#FFFFFF',
    'axes.grid': True, 'grid.alpha': 0.3,
    'axes.spines.top': False, 'axes.spines.right': False,
    'font.size': 11,
})
sns.set_palette('husl')

DATA_PATH = 'customers-1000000.csv'
print('  Libraries loaded & settings applied')
```

## Load Raw Data & First-Look Audit

```python
df_raw = pd.read_csv(DATA_PATH)

print('=' * 55)
print(f'  Shape          : {df_raw.shape[0]:,} rows × {df_raw.shape[1]} cols')
print(f'  Memory usage   : {df_raw.memory_usage(deep=True).sum() / 1e6:.1f} MB')
print('=' * 55)

print('\n Column dtypes:')
print(df_raw.dtypes.to_string())

print('\n Null counts:')
nulls = df_raw.isnull().sum()
print(nulls[nulls > 0] if nulls.sum() > 0 else '    No nulls found')

print(f'\n Full-row duplicates : {df_raw.duplicated().sum():,}')
print(f' Customer ID dupes   : {df_raw["Customer Id"].duplicated().sum():,}')

df_raw.head(3)
```

##  Column Names → snake_case

```python
df = df_raw.copy()

old_cols = df.columns.tolist()
df.columns = (
    df.columns
      .str.strip()
      .str.replace(r'[\s]+', '_', regex=True)
      .str.replace(r'[^\w]', '', regex=True)
      .str.lower()
)
new_cols = df.columns.tolist()

print(' Column rename map:')
for o, n in zip(old_cols, new_cols):
    marker = '→' if o != n else '='
    print(f'   {o:25s} {marker}  {n}')
```

##  Strip Whitespace (All String Columns)

```python
str_cols = df.select_dtypes('object').columns.tolist()

before_counts = {c: df[c].str.startswith(' ').sum() + df[c].str.endswith(' ').sum()
                 for c in str_cols}

df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())

total_fixed = sum(before_counts.values())
print(f' Whitespace stripped across {len(str_cols)} string columns')
print(f'   Total cells fixed: {total_fixed:,}')
for col, cnt in before_counts.items():
    if cnt > 0:
        print(f'   {col:25s}  →  {cnt:,} cells had leading/trailing spaces')
```

##  Duplicates (Full-Row + Customer ID)

```python
before = len(df)

# ① Full-row duplicates
full_dupes = df.duplicated().sum()
df.drop_duplicates(inplace=True)

# ② Customer ID duplicates (keep first)
id_dupes = df['customer_id'].duplicated().sum()
df.drop_duplicates(subset='customer_id', keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)

after = len(df)
print(f'  Duplicate removal complete')
print(f'   Full-row duplicates removed : {full_dupes:,}')
print(f'   Customer ID duplicates removed : {id_dupes:,}')
print(f'   Rows before: {before:,}  →  after: {after:,}')
```

## Customer ID → Uppercase + Validate Format

```python
df['customer_id'] = df['customer_id'].str.upper()

# Validate: must be alphanumeric, 10–20 chars (adjust regex to your real format)
id_pattern = r'^[A-Z0-9]{10,20}$'
invalid_ids = df[~df['customer_id'].str.match(id_pattern, na=False)]

print(f' Customer IDs uppercased')
print(f'   Total IDs      : {len(df):,}')
print(f'   Invalid format : {len(invalid_ids):,}')
if len(invalid_ids) > 0:
    print('   Sample invalid IDs:')
    print(invalid_ids['customer_id'].head(10).tolist())

print(f'\n   Sample IDs:')
print(df['customer_id'].head(5).tolist())
```

## Subscription Date → Datetime + Derived Features

```python
df['subscription_date'] = pd.to_datetime(df['subscription_date'], errors='coerce')

unparsed = df['subscription_date'].isna().sum()

# Derived time features
df['sub_year']         = df['subscription_date'].dt.year
df['sub_month']        = df['subscription_date'].dt.month
df['sub_quarter']      = df['subscription_date'].dt.quarter
df['sub_dow']          = df['subscription_date'].dt.day_name()
df['sub_is_weekend']   = df['subscription_date'].dt.weekday >= 5
df['tenure_days']      = (pd.Timestamp.today() - df['subscription_date']).dt.days
df['tenure_years']     = (df['tenure_days'] / 365.25).round(2)

print(f'  Date parsing complete')
print(f'   dtype          : {df["subscription_date"].dtype}')
print(f'   Unparseable    : {unparsed:,}')
print(f'   Date range     : {df["subscription_date"].min().date()} → {df["subscription_date"].max().date()}')
print(f'   Derived cols added: sub_year, sub_month, sub_quarter, sub_dow, sub_is_weekend, tenure_days, tenure_years')
print(f'\n   Weekend subs   : {df["sub_is_weekend"].sum():,} ({df["sub_is_weekend"].mean()*100:.1f}%)')
print(f'   Avg tenure     : {df["tenure_days"].mean():.0f} days  ({df["tenure_years"].mean():.1f} yrs)')
```

##  Phone Numbers → Standardized Format + Validation

```python
def clean_phone(p):
    """Normalize to (XXX) XXX-XXXX; tag invalid as INVALID_<raw>."""
    p = str(p).strip()
    # Strip extensions like x1234 or ext.123
    p = re.sub(r'(x|ext\.?)\d+$', '', p, flags=re.I).strip()
    # Keep digits only
    digits = re.sub(r'[^\d]', '', p)
    # Drop leading country code 1 for US numbers
    if len(digits) == 11 and digits[0] == '1':
        digits = digits[1:]
    if len(digits) == 10:
        return f'({digits[:3]}) {digits[3:6]}-{digits[6:]}'
    return f'INVALID_{p}'   # flag instead of silently corrupt

for col in ('phone_1', 'phone_2'):
    df[col] = df[col].apply(clean_phone)
    invalid = df[col].str.startswith('INVALID_').sum()
    valid_pattern = r'^\(\d{3}\) \d{3}-\d{4}$'
    valid = df[col].str.match(valid_pattern).sum()
    print(f'  {col}')
    print(f'   Valid   : {valid:,}  ({valid/len(df)*100:.2f}%)')
    print(f'   Invalid : {invalid:,}  ({invalid/len(df)*100:.2f}%)')
    if invalid > 0:
        print(f'   Samples : {df[df[col].str.startswith("INVALID_")][col].head(3).tolist()}')
    print()
```

##  Email → Validate + Extract Domain

```python
EMAIL_PATTERN = r'^[\w\.\+\-]+@[\w\-]+\.[a-zA-Z]{2,}$'

df['email'] = df['email'].str.lower().str.strip()
df['email_valid']  = df['email'].str.match(EMAIL_PATTERN, na=False)
df['email_domain'] = df['email'].str.split('@').str[1].str.lower()

invalid_emails = (~df['email_valid']).sum()
top_domains    = df['email_domain'].value_counts().head(10)

print(f' Email cleaning complete')
print(f'   Valid emails   : {df["email_valid"].sum():,}  ({df["email_valid"].mean()*100:.2f}%)')
print(f'   Invalid emails : {invalid_emails:,}')
if invalid_emails > 0:
    print('   Samples:', df[~df['email_valid']]['email'].head(5).tolist())
print(f'\n   Unique domains : {df["email_domain"].nunique():,}')
print(f'\n   Top 10 domains:')
print(top_domains.to_string())
```

##  Website → Force HTTPS + Validate URL

```python
# Force HTTPS
df['website'] = df['website'].str.replace(r'^http://', 'https://', regex=True)

# Validate basic URL structure
URL_PATTERN = r'^https?://[\w\-\.]+\.[a-zA-Z]{2,}(/.*)?$'
df['website_valid'] = df['website'].str.match(URL_PATTERN, na=False)

all_https  = df['website'].str.startswith('https').mean() * 100
valid_urls = df['website_valid'].sum()

print(f' Website cleaning complete')
print(f'   All HTTPS      : {all_https:.2f}%')
print(f'   Valid URLs     : {valid_urls:,}  ({valid_urls/len(df)*100:.2f}%)')
invalid_web = df[~df['website_valid']]
if len(invalid_web) > 0:
    print(f'   Invalid URLs   : {len(invalid_web):,}')
    print('   Samples:', invalid_web['website'].head(5).tolist())
```

##  Name Columns → Title Case + Check Empty

```python
name_cols = ['first_name', 'last_name', 'city', 'company']

for col in name_cols:
    before_sample = df[col].iloc[0]
    df[col] = df[col].str.strip().str.title()
    empty_count = df[col].isin(['', 'Nan', 'None']).sum() + df[col].isna().sum()
    print(f'  {col:15s}  |  sample: "{before_sample}" → "{df[col].iloc[0]}"  |  empty/null: {empty_count:,}')

# Full name as new convenience feature
df['full_name'] = df['first_name'] + ' ' + df['last_name']
print(f'\n   Added: full_name  →  sample: "{df["full_name"].iloc[0]}"')
```

## Country → Validate & Flag Unknowns

```python
df['country'] = df['country'].str.strip().str.title()

country_counts = df['country'].value_counts()
rare_countries  = country_counts[country_counts < 10]   # suspiciously rare → possible typos

print(f'  Country cleaning complete')
print(f'   Unique countries   : {df["country"].nunique():,}')
print(f'   Suspiciously rare  : {len(rare_countries):,}  (count < 10)')

if len(rare_countries) > 0:
    print('\n   Rare / possible typos:')
    print(rare_countries.to_string())

print('\n   Top 10 countries:')
print(country_counts.head(10).to_string())
```

## Final Schema & Quality Score

```python
# Per-column completeness
total = len(df)
quality = pd.DataFrame({
    'dtype'       : df.dtypes,
    'non_null'    : df.notnull().sum(),
    'null'        : df.isnull().sum(),
    'null_pct'    : (df.isnull().sum() / total * 100).round(2),
    'unique'      : df.nunique(),
    'unique_pct'  : (df.nunique() / total * 100).round(2),
})

overall_completeness = (1 - df.isnull().sum().sum() / (total * len(df.columns))) * 100

print('=' * 60)
print(f'   OVERALL DATA QUALITY SCORE : {overall_completeness:.2f}%')
print(f'  Rows: {total:,}  |  Columns: {len(df.columns)}')
print('=' * 60)
print()
print(quality.to_string())
```

##  EDA Visualizations (4-Panel)

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Customer Dataset — Key Insights', fontsize=16, fontweight='bold', y=1.01)

# ─── A: Monthly subscriptions trend ──────────────────────────────────
ax = axes[0, 0]
monthly = df.groupby(df['subscription_date'].dt.to_period('M')).size()
monthly.index = monthly.index.to_timestamp()
ax.plot(monthly.index, monthly.values, lw=2.5, color='#2980B9')
ax.fill_between(monthly.index, monthly.values, alpha=0.12, color='#2980B9')
peak_m, peak_v = monthly.idxmax(), monthly.max()
ax.annotate(f'Peak\n{peak_v:,}', xy=(peak_m, peak_v),
            xytext=(peak_m, peak_v + 350),
            arrowprops=dict(arrowstyle='->', color='#E74C3C', lw=1.5),
            fontsize=9, color='#E74C3C', ha='center')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{int(x):,}'))
ax.set_title(' Monthly Subscriptions', fontweight='bold')
ax.set_xlabel('Month'); ax.set_ylabel('New Subscribers')

# ─── B: Top 10 countries ─────────────────────────────────────────────
ax = axes[0, 1]
top_c = df['country'].value_counts().head(10)
colors = sns.color_palette('husl', 10)
bars = ax.barh(top_c.index[::-1], top_c.values[::-1], color=colors)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{int(x):,}'))
for bar in bars:
    w = bar.get_width()
    ax.text(w + 30, bar.get_y() + bar.get_height()/2, f'{int(w):,}',
            va='center', fontsize=8)
ax.set_title(' Top 10 Countries', fontweight='bold')
ax.set_xlabel('Customers')

# ─── C: Heatmap year × month ─────────────────────────────────────────
ax = axes[1, 0]
pivot = df.groupby(['sub_year','sub_month']).size().unstack(fill_value=0)
month_labels = ['Jan','Feb','Mar','Apr','May','Jun',
                'Jul','Aug','Sep','Oct','Nov','Dec']
sns.heatmap(pivot, annot=True, fmt=',d', cmap='YlOrRd',
            ax=ax, linewidths=0.4, cbar_kws={'shrink': 0.8},
            xticklabels=[month_labels[int(c)-1] for c in pivot.columns])
ax.set_title(' Subscription Heatmap  (Year × Month)', fontweight='bold')
ax.set_xlabel('Month'); ax.set_ylabel('Year')

# ─── D: Day-of-week bar ──────────────────────────────────────────────
ax = axes[1, 1]
dow_order  = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
dow_counts = df['sub_dow'].value_counts().reindex(dow_order)
bar_colors = ['#E74C3C' if d in ('Saturday','Sunday') else '#3498DB' for d in dow_order]
ax.bar(dow_counts.index, dow_counts.values, color=bar_colors)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{int(x):,}'))
ax.set_title(' Subscriptions by Day of Week\n(red = weekend)', fontweight='bold')
ax.set_ylabel('Subscribers')
ax.tick_params(axis='x', rotation=30)
for i, (val, lbl) in enumerate(zip(dow_counts.values, dow_order)):
    ax.text(i, val + 800, f'{int(val):,}', ha='center', fontsize=7.5)

plt.tight_layout()
plt.savefig('insights.png', dpi=150, bbox_inches='tight')
plt.show()
print(' Saved → insights.png')
```
